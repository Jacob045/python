```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import csv
filepath = r'I:\Data\Personal Data\graduation project\SACOL\original_data\200911\sounding_data_as_MWR_format_200911.txt'
```

```python
data = pd.read_csv(filepath,sep='  ')
photoname = filepath[68:102]
plt.rcParams['figure.figsize'] = (10.0,20.0)
#
Height = data.columns.values[1:48]
# 
Len = len(data['Time'])
i = 0
while i<Len:
    data_for_print = np.array(data.iloc[i:i+1,1:48]).T
    plt.plot(data_for_print, Height, label=str(data['Time'][i]))
    i = i+1
#X
x_ticks = np.linspace(180, 320, 8)
plt.xticks(x_ticks)
# 
plt.legend()
# 
plt.grid()
# 
plt.title(photoname,fontsize=20,fontweight='bold')
# 
plt.savefig(photoname,bbox_inches = 'tight')
# 
plt.show()
```
