### 此程序用于《毕业论文》3.1SOCAL站气象要素的季节变化特征分析

```python
import pandas as pd
import read_filepath_module
```

```python
path = r'I:\Data\Personal Data\graduation project\SACOL\microwave\2009'
target_str = 'lv2'
lv2_paths = read_filepath_module.read_filepath(path,target_str)
```

```python
def search_mean(path,drawing_data):
    data = pd.read_csv(path)
    data = data[data['10'] == 12]
    data.reset_index(inplace=True)
    fanwei = []
    for row in range(data.shape[0]):
        if('07' in data['Date/Time'][row][9:11]):
            fanwei.append(row)
        if('08' in data['Date/Time'][row][9:11]):
            fanwei.append(row)
    if(fanwei):
        drawing_data[path[69:79].replace('-','')] = data.iloc[fanwei[0]:fanwei[-1],11:58].mean(0)
```

```python
drawing_data = pd.DataFrame()
for lv2_path in lv2_paths:
    print(lv2_path)
    search_mean(lv2_path,drawing_data)
```

```python
drawing_data.to_csv(r'I:\Data\Personal Data\graduation project\SACOL\microwave'+'\\rh2009_Ti00.csv')
```
