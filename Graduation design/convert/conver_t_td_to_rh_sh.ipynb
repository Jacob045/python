## 函数声明
- 此程序将温度和露点温度转换为相对湿度和比湿
- 输入：t（温度）、td（露点温度）
- 输出：rh（relative humidity、相对湿度）、sh（specific humidity、比湿）
***
## 函数思路
- 本程序使用榆中站怀俄明格式数据作为源数据，提取标准等压面数据行，将温度和露点温度转换为相对湿度和比湿
- 大气学原理参考《大气物理学》盛裴轩著 2003年5月第1版

```python
import pandas as pd
import math
from tqdm import tqdm
```

```python
filepath = r'I:\Data\Personal Data\graduation project\yuzhongzhan\52983-1989-2021.FSL'
data = pd.read_table(filepath,sep='\s+',names=['0','1','2','3','4','5','6'])
# 0行数据的类型标号、1压力、2层中观测高度、3温度、4露点温度、5风向、6风速
# 新增加的列 7相对湿度、8比湿
data['7'] = ' '
data['8'] = ' '
data.head(10)
```

```python
def esat(T):
    return(6.1078*math.exp(17.2693882*(T-273.16)/(T-35.86)))
```

```python
def esit(T):
    return(6.1078*math.exp(21.8745584*(T-276.16)/(T-7.66)))
```

```python
for row in range(data.shape[0]):
    if((data.iloc[row,0] == 4) and (data.iloc[row,3] != '99999') and (data.iloc[row,4] != '99999')):
        # Wyoming数据是摄氏度，摄氏度变热力学温度
        T = float(data.iloc[row,3]) + 273.15
        Td = float(data.iloc[row,4]) + 273.15
        print(f'row = {row}, T is {round(T,2)} and Td is {round(Td,2)}')
        if(T <= -49.9+273.15): # 平液面
            E = esat(T) # 
            Es = esat(Td)
            print(f'平液面 E is {E} and Es is {Es}\n')
        else: # 平冰面
            E = esit(T)
            Es = esit(Td)
            print(f'平冰面 E is {E} and Es is {Es}\n')
        data.iloc[row,7] = 100*E/Es # 相对湿度是比湿/饱和比湿
```
